```python
import numpy 
from plotfunctions import plot_it
from neuralnetwork import NeuralNetwork
import matplotlib.pyplot as plt
# The following code is used for hiding the warnings and make this notebook clearer.
import warnings
warnings.filterwarnings('ignore')


numpy.random.seed(0)

# Initialize the NeuralNetwork with
# 2 input neurons
# 2 hidden neurons
# 1 output neuron
nn = NeuralNetwork([2,2,1])
print(nn.weights)
# Set the input data
X = numpy.array([[0, 0], [0, 1],
                [1, 0], [1, 1]])

# Set the labels, the correct results for the xor operation
y = numpy.array([0, 1, 
                 1, 0])
```

```python

# Call the fit function and train the network for a chosen number of epochs
nn.fit(X, y,epochs=10000)

plt.plot(range(1,len(nn.errors) + 1),nn.errors, marker='.')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('ErrorValue')
plt.show()
```

```python

# Show the prediction results
print("Final prediction")
#for s in X:
    #print(s, nn.predict_single_data_descriptive(s))
#print(nn.w000)
plot_it(X, y, nn)
#print(nn.weights)

```

```python

fig0=plt.figure()
fig0.show()
ax0=fig0.add_subplot(111)
x0=numpy.arange(0,10000,1000)
ax0.plot(x0,[nn.y01[i] for i in x0],c='b',marker=".",label='y01')
ax0.plot(x0,[nn.y02[i] for i in x0],c='m',marker=".",label='y02')
ax0.plot(x0,[nn.y11[i] for i in x0],c='k',marker=".",label='y11')
ax0.plot(x0,[nn.y12[i] for i in x0],c='g',marker=".",label='y12')
ax0.plot(x0,[nn.y2[i] for i in x0],c='r',marker=".",label='y2')
plt.legend(loc=2)
plt.draw()
plt.title('Activation Values')
plt.xlabel('Epochs')
plt.ylabel('Activation Value')
plt.show()
```

```python

fig=plt.figure()
fig.show()
ax=fig.add_subplot(111)
x=numpy.arange(10000)
ax.plot(x,nn.w000,c='b',marker=".",label='bias1')
ax.plot(x,nn.w001,c='g',marker=".",label='bias2')
ax.plot(x,nn.w010,c='c',marker=".",label='Weight1_1')
ax.plot(x,nn.w011,c='r',marker=".",label='Weight1_2')
ax.plot(x,nn.w020,c='m',marker=".",label='Weight2_1')
ax.plot(x,nn.w021,c='k',marker=".",label='Weight2_2')

plt.legend(loc=2)
plt.draw()
plt.title('First Layer Weights')
plt.xlabel('Epochs')
plt.ylabel('Weight Value')
plt.show()
```

```python

fig2=plt.figure()
fig2.show()
ax2=fig2.add_subplot(111)
x2=numpy.arange(10000)
ax2.plot(x2,nn.w100,c='b',marker=".",label='bias_f')
ax2.plot(x2,nn.w110,c='m',marker=".",label='Weight_1')
ax2.plot(x2,nn.w120,c='k',marker=".",label='Weight_2')

plt.legend(loc=2)
plt.draw()

plt.title('Last Layer Weights')
plt.xlabel('Epochs')
plt.ylabel('Weight Value')
plt.show()
```
